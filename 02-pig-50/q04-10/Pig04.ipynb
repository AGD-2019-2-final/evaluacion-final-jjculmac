```python
%%pig
--
-- Carga el archivo desde el disco duro
--
u = LOAD 'truck_event_text_partition.csv' USING PigStorage(',') AS (f1:INT, f2:INT, f3:CHARARRAY, f4:CHARARRAY, f5:INT, f6:INT, f7:CHARARRAY, f8:CHARARRAY, f9:CHARARRAY, f10:INT, f11:CHARARRAY, f12:CHARARRAY);
t = FOREACH u GENERATE $0,$1,$2;
y = LIMIT t 10;
s = ORDER y BY f1,f2,f3;
DUMP s;
STORE s INTO 'output';
fs -get output/ .;
```
